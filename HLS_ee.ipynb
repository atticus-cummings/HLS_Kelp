```python
import ee
import branca
import geemap
import matplotlib.pyplot as plt
ee.Authenticate()
ee.Initialize(project='ee-atticuscummings')
print(ee.String('Hello from the Earth Engine servers!').getInfo())
```

```python
image = ee.ImageCollection("NASA/HLS/HLSL30/v002")\
    .filterBounds(ee.Geometry.Point(-119.292, 34.2215)) \
    .filterDate('2020-01-01', '2020-12-31') \
    .median() \
    .clip(ee.Geometry.Point([-119.292, 34.2215]).buffer(10000))
visParams = {
    'bands': ['B4', 'B3', 'B2'],  # RGB bands for Landsat 8
    'min': 0,
    'max': 3000,
    'gamma': [1, 1, 1]
}     
```

```python
Map = geemap.Map()
Map.centerObject(ee.Geometry.Point([-122.292, 37.9015]), 10)
Map.addLayer(image, visParams, 'Landsat 8 image')

# Define the color palette
palette = ['#ffffcc', '#c2e699', '#78c679', '#31a354', '#006837']

# Add the colorbar using visParams and the palette
Map.add_colorbar(vis_params={'min': 0, 'max': 3000, 'palette': palette}, label="Reflectance")
Map

# If you want to use matplotlib to display a single image band
# Convert the image to a numpy array
region = ee.Geometry.Point([-122.292, 37.9015]).buffer(5000)
url = image.visualize(**visParams).getThumbURL({'region': region, 'dimensions': '512x512', 'format': 'png'})
img = plt.imread(url)
```
