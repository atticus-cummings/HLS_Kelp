```python
import os
import rasterio
import re
from rasterio.errors import RasterioIOError
import shutil
```

```python
folder = r'C:\Users\attic\HLS_Kelp\imagery\Isla_Vista_Kelp\10SGD'

desired_bands = ['B02','B03','B04','B05','B06','B07'] #2,3,4,5,6,7]
pattern = re.compile(r'\.(' + '|'.join(desired_bands) + r')\.tif$')

```

```python
def delete_folder(path):
    try:
        if not os.path.isdir(path):
            print(f"{path} is not a directory.")
            return
        
        for root, dirs, files in os.walk(path):
            if dirs:
                raise Exception(f"Folder {path} contains subfolders: {dirs}")
        files = os.listdir(path)

        for file in files:
            file_path = os.path.join(path, file)
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.remove(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)

        os.rmdir(path)
        print(f"Successfully removed: {path}")

    except Exception as e:
        print(f"Failed to remove: {path}. Error: {e}")
```
